## Статья с habr. Написана Хабр карьера + HR-платформа Beehive

```python
def clean_text(text):

    text = text.replace('\xa0', ' ')
    text = re.sub(r'[^\w\sа-яА-ЯёЁ-]', '', text)  # Оставляем только русские буквы, пробелы и дефисы
    
    return text
```

```python
def getTextByUrl(url):
    response = requests.get(url)

    html_content = response.text
    soup = BeautifulSoup(html_content, 'html.parser')
    
    # разбиение текста на параграфы
    paragraphs = soup.find_all('p')
    text_data = [p.get_text() for p in paragraphs]

    return text_data
```

```python
url = "https://habr.com/ru/specials/782048/" 

text_data = getTextByUrl(url)
processed_text = []
    
for text in text_data:
    # Убедимся, что абзац состоит не менее чем из 3 слов
     if len(text.split()) >= 3:
        processed_text.append(clean_text(text))
```

```python
processed_text
```

```python
len(processed_text)
```

```python
with open('habr.txt', 'w', encoding='utf-8') as f:
    for inner_list in processed_text:
        f.write(inner_list + '\n') 
```

## ИССЛЕДОВАНИЕ МЕТОДОВ И ИНСТРУМЕНТОВ ПРЕДОТВРАЩЕНИЯ ЭМОЦИОНАЛЬНОГО ВЫГОРАНИЯ У СОТРУДНИКОВ IT-КОМПАНИЙ

```python
import pdfplumber
import re
import pymorphy2
import nltk

def clean_text(text):
    cleaned_text = re.sub(r'\b[a-zA-Z]+\b', '', text)
    
    cleaned_text = re.sub(r'[0-9]', '', cleaned_text) 
    cleaned_text = re.sub(r'$$.*?$$', '', cleaned_text)
    cleaned_text = re.sub(r'_', '', cleaned_text)
    cleaned_text = re.sub(r'[^\w\sа-яА-ЯёЁ.]', '', cleaned_text)

    cleaned_text = re.sub(r'\s+', ' ', cleaned_text).strip()
    cleaned_text = re.sub(r"Московский экономический журнал", "", cleaned_text)
    cleaned_text = re.sub(r"Организационная психология. . Т. . .", "", cleaned_text)
    cleaned_text = re.sub(r"С. . . ОРГАНИЗАЦИОННАЯ ПСИХОЛОГИЯ ...", "", cleaned_text)
    cleaned_text = re.sub(r'\.\s*\.\s*', '', cleaned_text)
    cleaned_text = re.sub(r"Электронный научный журнал", "", cleaned_text)
    cleaned_text = re.sub(r"Гуманитарный научный журнал", "", cleaned_text)
    cleaned_text = re.sub(r"  часть", "", cleaned_text)

    return cleaned_text

with pdfplumber.open('issledovanie-metodov-i-instrumentov-predotvrascheniya-emotsionalnogo-vygoraniya-u-sotrudnikov-it-kompaniy.pdf') as pdf:
    result = []

    for page in pdf.pages:
        text = page.extract_text()
        if text:
            cleaned_text = clean_text(text)
            result.append(cleaned_text)
```

```python
result = result[1:-5]
```

```python
result[0]=result[0][105:]
result
```

```python
result[-1]=result[-1][:-230]
result
```

```python
with open('research1.txt', 'w', encoding='utf-8') as f:
    for inner_list in result:
        f.write(inner_list + '\n') 
```
