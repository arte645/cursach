## Статья с habr. Написана Хабр карьера + HR-платформа Beehive

```python
def clean_text(text):

    text = text.replace('\xa0', ' ')
    text = re.sub(r'[^\w\sа-яА-ЯёЁ-]', '', text)  # Оставляем только русские буквы, пробелы и дефисы
    text = text.lower()
    
    return text
```

```python
def preprocess_text(text):
    cleaned_text = clean_text(text)
    words = cleaned_text.split()
    morph = pymorphy2.MorphAnalyzer()
    stopwords = nltk.corpus.stopwords.words('russian')
    filtered_words = [morph.parse(word)[0].normal_form for word in words if word not in stopwords]
    
    return filtered_words
```

```python
def getTextByUrl(url):
    response = requests.get(url)

    html_content = response.text
    soup = BeautifulSoup(html_content, 'html.parser')
    
    # разбиение текста на параграфы
    paragraphs = soup.find_all('p')
    text_data = [p.get_text() for p in paragraphs]

    return text_data
```

```python
url = "https://habr.com/ru/specials/782048/" 

text_data = getTextByUrl(url)
processed_text = []
    
for text in text_data:
    # Убедимся, что абзац состоит не менее чем из 3 слов
     if len(text.split()) >= 3:
        processed_text.append(preprocess_text(text))
```

```python
processed_text
```

```python
len(processed_text)
```

```python
with open('habr.txt', 'w', encoding='utf-8') as f:
    for inner_list in processed_text:
        f.write(', '.join(map(str, inner_list)) + '\n') 
```

## ИССЛЕДОВАНИЕ МЕТОДОВ И ИНСТРУМЕНТОВ ПРЕДОТВРАЩЕНИЯ ЭМОЦИОНАЛЬНОГО ВЫГОРАНИЯ У СОТРУДНИКОВ IT-КОМПАНИЙ

```python
import pdfplumber
import re
import pymorphy2
import nltk

def clean_text(text):
    cleaned_text = re.sub(r'\b[a-zA-Z]+\b', '', text)
    
    cleaned_text = re.sub(r'[0-9]', '', cleaned_text) 
    cleaned_text = re.sub(r'$$.*?$$', '', cleaned_text)
    cleaned_text = re.sub(r'_', '', cleaned_text)
    cleaned_text = re.sub(r'[^\w\sа-яА-ЯёЁ]', '', cleaned_text)

    cleaned_text = re.sub(r'\s+', ' ', cleaned_text).strip()

    return cleaned_text

def preprocess_text(text):
    cleaned_text = clean_text(text)
    words = cleaned_text.split()
    
    morph = pymorphy2.MorphAnalyzer()
    stopwords = nltk.corpus.stopwords.words('russian')
    
    filtered_words = [morph.parse(word)[0].normal_form for word in words 
                      if word not in stopwords and len(word) >= 2]

    return filtered_words

with pdfplumber.open('issledovanie-metodov-i-instrumentov-predotvrascheniya-emotsionalnogo-vygoraniya-u-sotrudnikov-it-kompaniy.pdf') as pdf:
    result = []

    for page in pdf.pages:
        text = page.extract_text()
        if text:
            cleaned_text = preprocess_text(text)
            result.append(cleaned_text)
```

```python
result
```

```python
# для игнора вступления и конца
start_word = "введение"
end_word = 'список'
found = False

filtered_data = []
for inner_list in result:
    if end_word in inner_list:
        index = inner_list.index(end_word)
        filtered_data.append(inner_list[:index])
        break;
    if found:
        filtered_data.append(inner_list)
    elif start_word in inner_list:
        found = True
        index = inner_list.index(start_word)
        filtered_data.append(inner_list[index:])
```

```python
filtered_data
```

```python
with open('research1.txt', 'w', encoding='utf-8') as f:
    for inner_list in filtered_data:
        f.write(', '.join(map(str, inner_list)) + '\n') 
```
